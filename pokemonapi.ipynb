```python
# first separate by their type - characters - list
# pokedex? categories --> list of pokemons
```

```python
import requests
from IPython.display import clear_output as co
import pprint

class Pokemon:
    
    def __init__(self, name):
        self.name = name
        res = requests.get('https://pokeapi.co/api/v2/pokemon/{pokename}'.format(pokename=name))
        self.data = res.json()
        self.stats = {}
        
        types=[]
        
        for poketype in self.data['types']:
            types.append(poketype['type']['name'])
        self.stats['types'] = types
        
        self.stats['name'] = self.data['name']
        
        self.stats['height'] = self.data['height']
        
        self.stats['weight'] = self.data['weight']
        
        abilities = []
        
        for ability in self.data['abilities']:
            abilities.append(ability['ability']['name'])
        self.stats['ability'] = abilities
        
        statistics = {}
        
        for stattype in self.data['stats']:
            statistics[stattype['stat']['name']] = stattype['base_stat']
        self.stats['basestats'] = statistics
        

class Pokedex:
    
    @classmethod
    def run(self):
        print("Welcome to Py Pokedex!")
        
        pokemon_total_list = []
        type_dict = {}
        
        done = False
        while not done:
            confirmation = input("Would you like to continue? Y/N ")
            co()
            if confirmation == 'n':
                done = True
            elif confirmation == 'y':
                
                pokedex_quit = False
                while not pokedex_quit:
                    choice = input("What would you like to do? \n-Add Pokemon\n-Show Pokemon\n-Exit\n").lower()
                    co()
                    
                    if choice == 'exit':
                        pokedex_quit = True
                
                    elif choice == 'add':
                        pokemon_choice = input('Which Pokemon would you like to view? ').lower()
                        co()

                        poke=Pokemon(pokemon_choice)
                
                        pokemon_total_list.append(poke)
                
                        pprint.pprint('You have added the pokemon to your Pokedex: ' + pokemon_choice.title(), sort_dicts = False)

                    elif choice == 'show':
                        for pokemon in pokemon_total_list:
                            if pokemon.data['types'][0]['type']['name'] not in type_dict:
                                type_dict[pokemon.data['types'][0]['type']['name']] = [pokemon.stats]
                            else:
                                type_dict[pokemon.data['types'][0]['type']['name']].append(pokemon.stats)
                        pprint.pprint(type_dict, sort_dicts = False)
            
     


Pokedex.run()
```
